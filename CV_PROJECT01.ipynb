```python
# Îç∞Ïù¥ÌÑ∞Î•º ÏãúÍ∞ÅÌôîÌïòÍ∏∞ ÏúÑÌïú Ìï®Ïàò
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(10,10))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

# Î¨∏Ï†ú 1-2. Îã§ÏñëÌïú Î∞©Î≤ïÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Î™®Îç∏Ïùò ÏÑ±Îä• Ïò¨Î¶¨Í∏∞
hyperparameter Î≥ÄÍ≤Ω, Î™®Îç∏ ÏàòÏ†ï, optimizer Î≥ÄÍ≤Ω, skip connection, pre-trained model Îì± Îã§ÏñëÌïú Î∞©Î≤ïÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ ÎÜíÏùº Ïàò ÏûàÏäµÎãàÎã§. Îã§ÏñëÌïú Ïã§ÌóòÏùÑ ÌÜµÌï¥ ÏÑ±Îä•ÏùÑ Ïò¨Î†§ Î≥¥ÏÑ∏Ïöî.

```python
# Î¨∏Ï†ú 1-2. hyperparameter ÏÑ§Ï†ï
batch_size = 32
epochs = 5
IMG_HEIGHT = 256
IMG_WIDTH = 256
```

train Îç∞Ïù¥ÌÑ∞Í∞Ä 2Ï≤úÏû• Î∞ñÏóê ÎêòÏßÄ ÏïäÍ∏∞ ÎïåÎ¨∏Ïóê Data Augmentation(Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ï)ÏùÑ ÌÜµÌï¥ Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÎ•º ÎäòÎ¶ΩÎãàÎã§.
Data AugmentationÏùÄ Rescale, Rotation, Random Crop, Zoom, Flip Îì± Îã§ÏñëÌïú Í∏∞Î≤ïÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞Ïùò ÏñëÏùÑ Ï¶ùÍ∞ÄÏãúÌÇ§Îäî Í≤ÉÏûÖÎãàÎã§. rangeÏùò Ïà´ÏûêÎ•º Î≥ÄÌôîÏãúÌÇ§Í±∞ÎÇò FlipÏóêÏÑú True/FalseÎ•º Ï†ÅÏ†àÌïòÍ≤å ÏÇ¨Ïö©ÌïòÏó¨ Îã§ÏñëÌïú Ïã§ÌóòÏùÑ Ìï¥Î≥¥ÏÑ∏Ïöî.

```python
# Î¨∏Ï†ú 1-2. Îç∞Ïù¥ÌÑ∞ generator ÏÉùÏÑ±
# Training data generator
image_gen_train = ImageDataGenerator(rescale=1./255,
                                     rotation_range=45,
                                     width_shift_range=0.4,
                                     height_shift_range=0.4,
                                     zoom_range=0.4,
                                     shear_range=10,
                                     channel_shift_range=20,
                                     brightness_range=(0.8, 1.2),
                                     horizontal_flip=True,
                                     fill_mode='nearest',
                                     vertical_flip=True)

train_data_gen = image_gen_train.flow_from_directory(batch_size=batch_size,
                                                     directory=train_dir,
                                                     shuffle=True,
                                                     target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                     class_mode='binary')

# Validation data generator
image_gen_val = ImageDataGenerator(rescale=1./255)
val_data_gen = image_gen_val.flow_from_directory(batch_size=batch_size,
                                                 directory=validation_dir,
                                                 target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                 class_mode='binary')
```

```python
# Î¨∏Ï†ú 1-2. Î™®Îç∏ Íµ¨ÌòÑ
# VGG16 Î™®Îç∏ Íµ¨ÌòÑ
# Parameter Initialization ÏÑ§Ï†ï
kernel_initializer = tf.keras.initializers.he_normal()  # He Ï¥àÍ∏∞Ìôî

input_layer = tf.keras.layers.Input(shape=(256, 256, 3))

# Block 1
x = tf.keras.layers.Conv2D(64, (3, 3), padding='same', kernel_initializer=kernel_initializer)(input_layer)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Conv2D(64, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.MaxPool2D((2, 2))(x)

# Block 2
x = tf.keras.layers.Conv2D(128, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Conv2D(128, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.MaxPool2D((2, 2))(x)

# Block 3
x = tf.keras.layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.MaxPool2D((2, 2))(x)

# Block 4
x = tf.keras.layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.MaxPool2D((2, 2))(x)

# Block 5
x = tf.keras.layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=kernel_initializer)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.MaxPool2D((2, 2))(x)

# FC Layers
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(4096, activation='relu')(x)
x = tf.keras.layers.Dense(4096, activation='relu')(x)
x = tf.keras.layers.Dense(1000, activation='relu')(x)
# Output Layer
output_layer = tf.keras.layers.Dense(1, activation='sigmoid')(x)

# Create model
model = tf.keras.Model(inputs=[input_layer], outputs=[output_layer])
```

```python
# Î¨∏Ï†ú 1-2. loss function, optimizer, metric ÏÑ§Ï†ï Î∞è Î™®Îç∏ Ïª¥ÌååÏùº
loss_function=tf.keras.losses.binary_crossentropy
optimize=tf.keras.optimizers.Adam(learning_rate=0.01)
metric=tf.keras.metrics.binary_accuracy
model.compile(loss=loss_function,
              optimizer=optimize,
              metrics=[metric])

callbacks_list= [tf.keras.callbacks.TensorBoard(log_dir='log_dir', histogram_freq=1)]
# callback Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÍ≥† Ïã∂Îã§Î©¥ Ï∂îÍ∞ÄÌï¥ÏÑú ÌïôÏäµÌïòÎäî Îç∞Ïóê ÌôúÏö©Ìï¥ Î≥¥ÏÑ∏Ïöî.
# Î™®Îç∏ ÌïôÏäµÏùÑ ÏãúÏºúÎ¥ÖÏãúÎã§.
# data generatorÎäî ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ÏôÄ ÌÉÄÍ≤ü(ÎùºÎ≤®)Ïùò batchÎ•º ÎÅùÏóÜÏù¥ Î∞òÌôòÌï©ÎãàÎã§.
# batchÍ∞Ä ÎÅùÏóÜÏù¥ ÏÉùÏÑ±ÎêòÍ∏∞ ÎïåÎ¨∏Ïóê, Ìïú Î≤àÏùò epochÏóê generatorÎ°úÎ∂ÄÌÑ∞ ÏñºÎßàÎÇò ÎßéÏùÄ ÏÉòÌîåÏùÑ ÎΩëÏùÑÏßÄ Î™®Îç∏Ïóê Ï†ÑÎã¨Ìï¥Ïïº Ìï©ÎãàÎã§.
# ÎßåÏïΩ batch_size=20Ïù¥Í≥† steps_per_epoch=100Ïùº Í≤ΩÏö∞ (Îç∞Ïù¥ÌÑ∞, ÎùºÎ≤®)Ïùò Ïåç 20Í∞úÍ∞Ä ÏÉùÏÑ±ÎêòÍ≥†, ÌÅ¨Í∏∞Í∞Ä 20Ïù∏ batch Îç∞Ïù¥ÌÑ∞Î•º 100Î≤à ÌïôÏäµÌïòÎ©¥ 1 epochÏù¥ ÏôÑÎ£åÎê©ÎãàÎã§. Îã®, ÌÅ¨Í∏∞ 20Ïùò batch Îç∞Ïù¥ÌÑ∞Îäî Îß§Î≤à ÎûúÎç§ÏúºÎ°ú ÏÉùÏÑ±Îê©ÎãàÎã§.

# -- ÏùºÎ∞òÏ†ÅÏúºÎ°ú (Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ Í∏∏Ïù¥/batch_size)Î•º steps_per_epochÏúºÎ°ú ÏÑ§Ï†ïÌï©ÎãàÎã§.
```

```python
# Î™®Îç∏ ÌïôÏäµ
history = model.fit(
      train_data_gen,
      steps_per_epoch=int((len(os.listdir(train_cats_dir)) + len(os.listdir(train_dogs_dir)))/batch_size),
      epochs=epochs,
      validation_data=val_data_gen,
      # callbacks=callbacks_list,
      validation_freq=1)
```
